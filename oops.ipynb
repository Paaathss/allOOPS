ENCAPSULATION

Encapsulation binds together data and functions that manipulate data,
keeping it safe from inheritance and missuse .

encapsulation is the one of fundamental concepts in obj oriented prgm .

it describes the idea of wrapping data and methods that works on data within one unit .

this puts restrictions on accessing variables and methods directly and can prevent accidental modification of data .

To prevent accidandal change  , an object variable can only be changed by an object method . those type of variables are known as private variables

self.name : Public
self._name :protected -- private
self.__name :private -- strongly private

```python
class Encapsulation():
  def __init__(self,a,b,c):
    self.a = a
    self._b = b
    self.__c = c

  def mymethod(self):
    print(self.__c)
```

```python
x=Encapsulation(2,3,4)
print(x.a)
print(x._b)
x.mymethod()
```

#ABSTRACTION METHOD

Abstraction is used to hide the internal functionality of the function from the users .

The users only interact with the basuc implementation of the function, but the inner working is hidden .

```python
from abc import ABC,abstractmethod

class Mail(ABC):

  @abstractmethod
  def mileage(self):
    pass
  def fun(self):
    print('fun')

class Child(Mail):

  def sum(self):
    print('sum')

  def mileage(self):
    print('mileage')

obj = Child()
obj.mileage()
```
